```python
import numpy as np
import pandas as pd
```

```python
books = pd.read_csv('/content/Books.csv' , names=['ISBN', 'Book_Title', 'Book_Author' , 'Year_of_Publication' , 'Publisher' , 'Image_URL_S','Image_URL_M','Image_URL_L'], header=None)
ratings = pd.read_csv('/content/Ratings.csv',names=['User_ID','ISBN','Book_Rating'],header=None)
users = pd.read_csv('/content/Users.csv')
```

```python
print(books.shape)
print(ratings.shape)
print(users.shape)
```

```python
books.isnull().sum()
```

```python
books.dropna(inplace=True)
books.isnull().sum()
```

```python
ratings.isnull().sum()
```

```python
books.sample(1)
```

```python
ratings.sample(1)
```

```python
books = books.drop(['Image_URL_S' , 'Image_URL_L' ,''], axis=1, errors='ignore')
books.shape
```

```python
books.sample(1)
```

```python
book_data = ratings.merge(books, on='ISBN')
book_data['Image_URL_M'] = book_data['Image_URL_M'].str.replace(
    'http://', 'https://', regex=False
)
book_data.sample(5)
```

```python
book_data.shape
```

```python
num_rating_df = book_data.groupby('Book_Title').count()['Book_Rating'].reset_index()
num_rating_df.rename(columns={'Book_Rating': 'num_ratings'}, inplace=True)
num_rating_df
```

```python
book_data['Book_Rating'] = pd.to_numeric(book_data['Book_Rating'], errors='coerce')
avg_rating_df = book_data.groupby('Book_Title')['Book_Rating'].mean().reset_index()
avg_rating_df.rename(columns={'Book_Rating': 'avg_ratings'}, inplace=True)
avg_rating_df
```

```python
popular_df = num_rating_df.merge(avg_rating_df, on='Book_Title')
popular_df
```

```python
popular_df = popular_df[popular_df['num_ratings'] >= 250].sort_values('avg_ratings', ascending=False)
popular_df
```

```python
books_data = popular_df.merge(books, on='Book_Title').drop_duplicates('Book_Title')[['Book_Title', 'Book_Author', 'num_ratings', 'avg_ratings',	'Image_URL_M']]
books_data
```

```python
x = book_data.groupby('User_ID').count()['Book_Rating'] > 200
users_data = x[x].index
filtered_rating = book_data[book_data['User_ID'].isin(users_data)]
filtered_rating
```

```python
y = filtered_rating.groupby('Book_Title').count()['Book_Rating'] >= 50
famous_books = y[y].index
```

```python
final_ratings = filtered_rating[filtered_rating['Book_Title'].isin(famous_books)]
final_ratings
```

```python
pt = final_ratings.pivot_table(index='Book_Title', columns='User_ID', values='Book_Rating')
pt.fillna(0, inplace=True)
pt.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity_scores = cosine_similarity(pt)
similarity_scores.shape
```

```python
# recommender system
def recommend(book_name):
  index  = np.where(pt.index == book_name)[0][0]
  similar_items = sorted(list(enumerate(similarity_scores[index])),key = lambda x:x[1], reverse=True)[1:5]

  data = []
  for i in similar_items:
    item = []
    temp_df = books[books['Book_Title'] == pt.index[i[0]]]
    item.extend(list(temp_df.drop_duplicates('Book_Title')['Book_Title'].values))
    item.extend(list(temp_df.drop_duplicates('Book_Title')['Book_Author'].values))
    item.extend(list(temp_df.drop_duplicates('Book_Title')['Image_URL_M'].values))
    data.append(item)

  return data
```

```python
recommend('1984')
```
